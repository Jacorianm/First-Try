```python
import datetime
from google.colab import output
import IPython

# ==========================================
# 1. THE WEBSITE DESIGN (HTML & JAVASCRIPT)
# ==========================================
html_content = """
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Marcell | IT Infrastructure Hub</title>
    <link href="https://cdn.jsdelivr.net/npm/bootstrap@5.3.0/dist/css/bootstrap.min.css" rel="stylesheet">
    <link rel="stylesheet" href="https://cdnjs.cloudflare.com/ajax/libs/font-awesome/6.4.0/css/all.min.css">
    <style>
        body { background-color: #0f172a; color: #e2e8f0; font-family: 'Segoe UI', sans-serif; }
        .hero-section { background: linear-gradient(135deg, #1e1b4b 0%, #311042 100%); border-bottom: 2px solid #4338ca; padding: 40px 0; position: relative; }
        .card { background-color: #1e293b; border: 1px solid #334155; border-radius: 12px; }
        .form-control { background-color: #0f172a; border: 1px solid #475569; color: #fff; }
        .form-control:focus { background-color: #0f172a; color: #fff; border-color: #6366f1; box-shadow: none; }
        .text-gradient { background: linear-gradient(to right, #818cf8, #c084fc); -webkit-background-clip: text; -webkit-text-fill-color: transparent; }

        /* Floating Gear Button & Dropdown positioning */
        .admin-gear-dropdown { position: absolute; top: 20px; right: 20px; z-index: 1000; }
        .gear-btn { background: none; border: none; color: #94a3b8; font-size: 24px; transition: color 0.3s, transform 0.3s; }
        .gear-btn:hover { color: #eab308; transform: rotate(45deg); }
        .dropdown-menu { background-color: #1e293b; border: 1px solid #475569; }
        .dropdown-item { color: #e2e8f0; }
        .dropdown-item:hover { background-color: #334155; color: #fff; }
    </style>
</head>
<body>

    <header class="hero-section text-center">
        <div class="dropdown admin-gear-dropdown">
            <button class="gear-btn" type="button" data-bs-toggle="dropdown" aria-expanded="false">
                <i class="fas fa-cog"></i>
            </button>
            <ul class="dropdown-menu dropdown-menu-end shadow">
                <li><a class="dropdown-item" href="#" onclick="alert('System Optimization: 100%\\nAll network metrics nominal.')"><i class="fas fa-sliders-h me-2"></i> Settings</a></li>
                <li><hr class="dropdown-divider border-secondary"></li>
                <li><a class="dropdown-item text-warning" href="#" data-bs-toggle="modal" data-bs-target="#loginModal"><i class="fas fa-lock me-2"></i> Admin Login</a></li>
            </ul>
        </div>

        <div class="container">
            <h1 class="display-4 fw-bold text-gradient">Marcell's IT Infrastructure Hub</h1>
            <p class="lead text-muted">Network Portal & API Request Monitoring Center</p>
        </div>
    </header>

    <main class="container my-5">
        <div id="mainHub" class="row justify-content-center">
            <div class="col-md-6">
                <div class="card p-4 text-center">
                    <h3 class="mb-3 text-info"><i class="fas fa-users me-2"></i> Guest Hub Terminal</h3>
                    <p class="text-muted small">Enter your identifier to authenticate your network packet and pass through to the main network array.</p>

                    <div class="mb-3 text-start">
                        <label class="form-label text-muted">User Gateway Identifier</label>
                        <input type="text" id="visitorName" class="form-control" placeholder="Enter name or guest ID...">
                    </div>
                    <button onclick="navigateToBlankPage()" class="btn btn-info w-100 fw-bold text-dark">Enter Guest Hub</button>
                </div>
            </div>
        </div>

        <div id="blankPage" class="row justify-content-center d-none">
            <div class="col-md-8 text-center">
                <div class="card p-5 border-info">
                    <h2 class="text-info mb-3">🌌 Welcome to the Internal Hub</h2>
                    <p class="text-muted">You have successfully triggered a network API tracking block. Your node address has been verified and registered.</p>
                    <div class="p-4 bg-dark bg-opacity-40 rounded border border-secondary my-4">
                        <p class="mb-0 text-muted font-monospace">// This is a placeholder space for your upcoming skills page deployment.</p>
                    </div>
                    <button onclick="returnToHub()" class="btn btn-outline-info btn-sm">Disconnect Framework & Return Home</button>
                </div>
            </div>
        </div>

        <div id="adminDashboard" class="row d-none">
            <div class="col-12">
                <div class="card p-4 border-warning">
                    <div class="d-flex justify-content-between align-items-center border-bottom border-secondary pb-3 mb-3">
                        <h3 class="text-warning mb-0"><i class="fas fa-terminal me-2"></i> Admin Central Logging Console</h3>
                        <div>
                            <button onclick="refreshLogs()" class="btn btn-sm btn-outline-warning me-2">Refresh Bus Stream</button>
                            <button onclick="logoutAdmin()" class="btn btn-sm btn-outline-danger">Disconnect Session</button>
                        </div>
                    </div>
                    <pre id="logDisplay" class="p-3 rounded border border-secondary" style="background-color: #090d16; color: #38bdf8; max-height: 400px; overflow-y: auto; font-family: monospace;">No system telemetry records found.</pre>
                </div>
            </div>
        </div>
    </main>

    <div class="modal fade" id="loginModal" tabindex="-1" aria-hidden="true">
        <div class="modal-dialog modal-dialog-centered">
            <div class="modal-content card p-2">
                <div class="modal-header border-0">
                    <h5 class="modal-title text-warning"><i class="fas fa-shield-alt me-2"></i> Security Encryption Gate</h5>
                    <button type="button" class="btn-close btn-close-white" data-bs-dismiss="modal" aria-label="Close"></button>
                </div>
                <div class="modal-body">
                    <div class="mb-3">
                        <label class="form-label text-muted">Security Encryption Key</label>
                        <input type="password" id="adminPassword" class="form-control" placeholder="••••••••">
                    </div>
                    <div id="adminStatus" class="alert d-none mb-0"></div>
                </div>
                <div class="modal-footer border-0">
                    <button type="button" class="btn btn-outline-secondary" data-bs-dismiss="modal">Cancel</button>
                    <button type="button" onclick="checkAdminLogin()" class="btn btn-warning fw-bold text-dark">Authenticate</button>
                </div>
            </div>
        </div>
    </div>

    <script src="https://cdn.jsdelivr.net/npm/bootstrap@5.3.0/dist/js/bootstrap.bundle.min.js"></script>

    <script>
        // Track the user name and flip views to the blank page
        function navigateToBlankPage() {
            const name = document.getElementById('visitorName').value || 'Anonymous Node';

            // Hide main hub, show our empty page layout
            document.getElementById('mainHub').classList.add('d-none');
            document.getElementById('blankPage').classList.remove('d-none');

            // Dispatch tracking packet to Python tracker backend
            google.colab.kernel.invokeFunction('notebook.log_visit', [name], {});
        }

        function returnToHub() {
            document.getElementById('visitorName').value = '';
            document.getElementById('blankPage').classList.add('d-none');
            document.getElementById('mainHub').classList.remove('d-none');
        }

        function checkAdminLogin() {
            const password = document.getElementById('adminPassword').value;
            google.colab.kernel.invokeFunction('notebook.admin_login', [password], {});
        }

        function showAdminDashboard(logsJson) {
            // Close modal manually via Bootstrap API
            const modalEl = document.getElementById('loginModal');
            const modal = bootstrap.Modal.getInstance(modalEl);
            if(modal) modal.hide();

            // Switch layout views to dashboard panel
            document.getElementById('mainHub').classList.add('d-none');
            document.getElementById('blankPage').classList.add('d-none');
            document.getElementById('adminDashboard').classList.remove('d-none');
            document.getElementById('logDisplay').innerText = JSON.stringify(logsJson, null, 2);
        }

        function denyAdminAccess() {
            const errDiv = document.getElementById('adminStatus');
            errDiv.className = "alert alert-danger bg-danger bg-opacity-10 text-danger border-danger mt-2";
            errDiv.innerText = "CRITICAL ERROR: Authentication Key Denied.";
            errDiv.classList.remove('d-none');
        }

        function logoutAdmin() {
            document.getElementById('adminPassword').value = '';
            document.getElementById('adminStatus').classList.add('d-none');
            document.getElementById('adminDashboard').classList.add('d-none');
            document.getElementById('mainHub').classList.remove('d-none');
        }

        function refreshLogs() {
            google.colab.kernel.invokeFunction('notebook.get_admin_logs', [], {});
        }
    </script>
</body>
</html>
"""

with open('index.html', 'w') as f:
    f.write(html_content)

# ==========================================
# 2. THE BACKEND TRACKER LOGIC (PYTHON)
# ==========================================
class APITracker:
    def __init__(self):
        self.calls = []

    def log_request(self, endpoint, method, payload=None):
        request_data = {
            "timestamp": datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S"),
            "type": "INBOUND_REQUEST",
            "endpoint": endpoint,
            "method": method,
            "payload": payload
        }
        self.calls.append(request_data)
        return request_data

    def log_response(self, request_data, status_code, response_body=None):
        response_data = {
            "timestamp": datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S"),
            "type": "OUTBOUND_RESPONSE",
            "request_id": request_data.get("timestamp"),
            "status_code": status_code,
            "response_body": response_body,
        }
        self.calls.append(response_data)

# Boot track stream
tracker = APITracker()
ADMIN_PASSWORD = "Zillathepope"

def log_visit(name):
    # Logs the exact user login name directly inside the tracking records database array
    req_log = tracker.log_request("/hub/entry", "POST", payload={"tracked_username": name})
    tracker.log_response(req_log, 200, response_body={"status": "session_captured", "user": name})

def admin_login(password):
    if password == ADMIN_PASSWORD:
        output.eval_js(f"showAdminDashboard({tracker.calls})")
    else:
        output.eval_js("denyAdminAccess()")

def get_admin_logs():
    output.eval_js(f"document.getElementById('logDisplay').innerText = JSON.stringify({tracker.calls}, null, 2)")

# Connect Frontend to Backend bridges
output.register_callback('notebook.log_visit', log_visit)
output.register_callback('notebook.admin_login', admin_login)
output.register_callback('notebook.get_admin_logs', get_admin_logs)

# ==========================================
# 3. LAUNCH INTERFACE
# ==========================================
print("Icon updated. View layout rendered directly below:")
IPython.display.HTML(filename='index.html')
```

```python
import datetime
from google.colab import output
import IPython

# ==========================================
# 1. THE WEBSITE DESIGN (HTML & JAVASCRIPT)
# ==========================================
html_content = """
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Marcell's Lab</title>
    <link href="https://cdn.jsdelivr.net/npm/bootstrap@5.3.0/dist/css/bootstrap.min.css" rel="stylesheet">
    <link rel="stylesheet" href="https://cdnjs.cloudflare.com/ajax/libs/font-awesome/6.4.0/css/all.min.css">
    <style>
        /* CSS VARIABLES FOR THEME SWAPPING */
        :root {
            --theme-color: #00ffff; /* Default: Cyan/Aqua for Cyberpunk */
            --secondary-color: #ff00ff; /* Magenta accent */
            --glow-color: rgba(0, 255, 255, 0.6); /* Cyan glow */
            --dark-bg: #0d0d1a; /* Very dark background */
            --dark-card-bg: #1a1a2e; /* Slightly lighter dark for cards */
            --border-color: #33334a;
        }

        /* CYBERPUNK 2077 COLOR PALETTE & FONTS */
        body {
            background-color: var(--dark-bg);
            color: var(--theme-color);
            font-family: 'Consolas', 'Courier New', Courier, monospace; /* Techy font */
            text-shadow: 0 0 5px var(--glow-color);
            transition: filter 0.3s ease, background-color 0.3s ease;
            cursor: url('data:image/svg+xml;utf8,<svg xmlns="http://www.w3.org/2000/svg" width="24" height="24" viewBox="0 0 24 24"><path fill="%2300FFFF" d="M12 2C6.48 2 2 6.48 2 12s4.48 10 10 10 10-4.48 10-10S17.52 2 12 2zm0 18c-4.41 0-8-3.59-8-8s3.59-8 8-8 8 3.59 8 8-3.59 8-8 8zM11 7h2v6h-2zm0 8h2v2h-2z"/></svg>') 12 12, auto; /* Custom cursor */
        }

        /* Negative View Inversion (keep functional) */
        body.negative-mode {
            filter: invert(1);
            background-color: #ffffff; /* Forces clean white bg when inverted */
            color: #000000;
            text-shadow: none;
        }

        .hero-section {
            background: linear-gradient(135deg, #1f005c 0%, #000000 50%, #4b0082 100%); /* Dark gradient with more colors */
            border-bottom: 2px solid var(--theme-color);
            padding: 60px 0;
            position: relative;
            box-shadow: 0 0 20px var(--glow-color);
        }

        .card {
            background-color: var(--dark-card-bg);
            border: 1px solid var(--theme-color);
            border-radius: 5px;
            box-shadow: 0 0 15px var(--glow-color);
        }

        /* Input Fields */
        .form-control {
            background-color: #0d0d1a;
            border: 1px solid var(--border-color);
            color: var(--theme-color) !important;
            border-radius: 3px;
            font-family: 'Consolas', 'Courier New', Courier, monospace;
            text-shadow: none;
        }
        .form-control:focus {
            background-color: #0d0d1a;
            color: var(--theme-color) !important;
            border-color: var(--theme-color);
            box-shadow: 0 0 10px var(--glow-color);
        }
        .form-control::placeholder { color: rgba(0, 255, 255, 0.5) !important; opacity: 1; text-shadow: none; }
        .text-pure-white { color: #ffffff !important; text-shadow: 0 0 5px rgba(255, 255, 255, 0.4); font-weight: bold; }

        /* Text Colors */
        .text-info { color: var(--theme-color) !important; text-shadow: 0 0 5px var(--glow-color); }
        .text-warning { color: var(--secondary-color) !important; text-shadow: 0 0 5px rgba(255, 0, 255, 0.6); }
        .text-danger { color: #ff3366 !important; text-shadow: 0 0 5px rgba(255, 51, 102, 0.6); }
        .text-muted { color: rgba(255, 255, 255, 0.6) !important; text-shadow: none; }
        .text-gradient {
            background: linear-gradient(to right, var(--theme-color), var(--secondary-color));
            -webkit-background-clip: text;
            -webkit-text-fill-color: transparent;
            text-shadow: none;
        }

        /* Cyberpunk Buttons */
        .btn {
            border-radius: 3px;
            text-transform: uppercase;
            font-weight: bold;
            letter-spacing: 1px;
            transition: all 0.2s ease-in-out;
            text-shadow: none; /* Remove default text shadow from buttons */
        }
        .btn-info, .btn-outline-info {
            background-color: transparent;
            color: var(--theme-color) !important;
            border: 1px solid var(--theme-color);
            box-shadow: 0 0 5px var(--glow-color);
        }
        .btn-info:hover, .btn-outline-info:hover {
            background-color: var(--theme-color);
            color: #000 !important;
            box-shadow: 0 0 20px var(--glow-color);
        }
        .btn-warning, .btn-outline-warning {
            background-color: transparent;
            color: var(--secondary-color) !important;
            border: 1px solid var(--secondary-color);
            box-shadow: 0 0 5px rgba(255, 0, 255, 0.6);
        }
        .btn-warning:hover, .btn-outline-warning:hover {
            background-color: var(--secondary-color);
            color: #000 !important;
            box-shadow: 0 0 20px rgba(255, 0, 255, 0.6);
        }
        .btn-outline-danger {
            background-color: transparent;
            color: #ff3366 !important;
            border: 1px solid #ff3366;
            box-shadow: 0 0 5px rgba(255, 51, 102, 0.6);
        }
        .btn-outline-danger:hover {
            background-color: #ff3366;
            color: #000 !important;
            box-shadow: 0 0 20px rgba(255, 51, 102, 0.6);
        }
        .btn-outline-secondary {
            background-color: transparent;
            color: #ffffff !important;
            border: 1px solid #ffffff;
            box-shadow: 0 0 5px rgba(255, 255, 255, 0.4);
        }
        .btn-outline-secondary:hover {
            background-color: #ffffff;
            color: #000 !important;
            box-shadow: 0 0 20px rgba(255, 255, 255, 0.4);
        }

        /* Floating Gear Button & Dropdown */
        .admin-gear-dropdown { position: absolute; top: 20px; right: 20px; z-index: 1000; }
        .gear-btn {
            background: none;
            border: none;
            color: var(--theme-color);
            font-size: 28px;
            transition: color 0.3s, transform 0.3s, text-shadow 0.3s;
            text-shadow: 0 0 5px var(--glow-color);
        }
        .gear-btn:hover {
            color: #ffffff;
            transform: rotate(720deg); /* Full rotation */
            text-shadow: 0 0 20px #ffffff;
        }
        .dropdown-menu { background-color: var(--dark-card-bg); border: 1px solid var(--theme-color); border-radius: 5px; box-shadow: 0 0 10px var(--glow-color); }
        .dropdown-item { color: var(--theme-color); font-family: 'Consolas', 'Courier New', Courier, monospace; text-shadow: none; }
        .dropdown-item:hover { background-color: var(--theme-color); color: #000 !important; text-shadow: none; }
        .dropdown-divider { border-color: var(--border-color); opacity: 0.6; }

        /* Modal / Popups */
        .modal-content { background-color: var(--dark-card-bg); border: 1px solid var(--theme-color); border-radius: 8px; box-shadow: 0 0 30px var(--glow-color); }
        .modal-header { border-bottom: 1px solid var(--border-color); }
        .modal-footer { border-top: 1px solid var(--border-color); }
        .btn-close { filter: invert(1) grayscale(100%) brightness(200%); opacity: 0.8; }
        .btn-close:hover { opacity: 1; }

        /* Alerts & Preformatted Logs */
        pre#logDisplay {
            background-color: #0d0d1a !important;
            color: var(--theme-color) !important;
            border: 1px solid var(--border-color) !important;
            border-radius: 5px;
            text-shadow: 0 0 3px var(--glow-color);
            white-space: pre-wrap; /* Ensure text wraps */
            word-wrap: break-word; /* Break long words */
        }
        .alert {
            background-color: rgba(0, 255, 255, 0.1) !important; /* Semi-transparent glow */
            border: 1px solid var(--theme-color) !important;
            color: var(--theme-color) !important;
            border-radius: 5px;
            text-shadow: 0 0 3px var(--glow-color);
        }

        /* Form switch for Negative View styling */
        .form-check-input { background-color: var(--border-color); border: 1px solid var(--theme-color); border-radius: 50% !important; /* Rounded switch */}
        .form-check-input:checked { background-color: var(--theme-color); border-color: var(--theme-color); box-shadow: 0 0 5px var(--glow-color); }
        .form-check-label { text-shadow: none; }

        /* Footer */
        footer {
            color: rgba(255, 255, 255, 0.4) !important;
            text-shadow: none;
            border-color: var(--border-color) !important;
            padding-top: 20px !important;
            margin-top: 40px !important;
        }

        /* Animation for page transition */
        .page-transition {
            opacity: 0;
            transition: opacity 0.5s ease-in-out;
        }
        .page-transition.active {
            opacity: 1;
        }

        /* Nav Tabs for Admin Panel */
        .nav-tabs {
            border-bottom: 1px solid var(--border-color);
        }
        .nav-tabs .nav-link {
            color: var(--theme-color);
            border: 1px solid transparent;
            border-top-left-radius: 0.25rem;
            border-top-right-radius: 0.25rem;
        }
        .nav-tabs .nav-link:hover,
        .nav-tabs .nav-link.active {
            color: #000;
            background-color: var(--theme-color);
            border-color: var(--theme-color);
            border-bottom-color: var(--theme-color);
            box-shadow: 0 0 10px var(--glow-color);
        }
        .nav-tabs .nav-link.active {
            font-weight: bold;
        }

        /* Gallery Image Style */
        .gallery-image-preview {
            width: 100px;
            height: 100px;
            object-fit: cover;
            border: 1px solid var(--border-color);
            margin: 5px;
            box-shadow: 0 0 5px var(--glow-color);
        }

        /* Spark Effect */
        .spark {
            position: absolute;
            background-color: var(--theme-color); /* Use theme color for spark */
            border-radius: 50%;
            opacity: 0;
            transform: scale(0);
            animation: spark-animation 0.7s ease-out forwards;
            pointer-events: none;
            z-index: 9999;
        }

        @keyframes spark-animation {
            0% {
                opacity: 1;
                transform: scale(0.1);
                width: 10px;
                height: 10px;
                box-shadow: 0 0 5px var(--glow-color); /* Initial glow */
            }
            25% {
                opacity: 0.9;
                transform: scale(1.2);
                width: 15px;
                height: 15px;
                box-shadow: 0 0 10px var(--glow-color), 0 0 20px var(--secondary-color); /* Burst glow */
            }
            50% {
                opacity: 0.6;
                transform: scale(1.8);
                width: 25px;
                height: 25px;
                box-shadow: 0 0 15px var(--glow-color), 0 0 30px var(--secondary-color); /* Peak burst glow */
            }
            100% {
                opacity: 0;
                transform: scale(2.5);
                width: 35px;
                height: 35px;
                box-shadow: 0 0 0px var(--glow-color); /* Fade out glow */
            }
        }
    </style>
</head>
<body>

    <header class="hero-section text-center text-white">
        <div class="dropdown admin-gear-dropdown">
            <button class="gear-btn" type="button" data-bs-toggle="dropdown" aria-expanded="false">
                <i class="fas fa-cog"></i>
            </button>
            <ul class="dropdown-menu dropdown-menu-end shadow">
                <li><a class="dropdown-item" href="#" data-bs-toggle="modal" data-bs-target="#settingsModal"><i class="fas fa-sliders-h me-2"></i> System Settings</a></li>
                <li><hr class="dropdown-divider"></li>
                <li><a class="dropdown-item" href="#" data-bs-toggle="modal" data-bs-target="#loginModal"><i class="fas fa-lock me-2"></i> Admin Login</a></li>
            </ul>
        </div>

        <div class="container">
            <h1 class="display-4 fw-bold text-gradient">Marcell's Lab</h1>
            <p class="lead text-muted">Decentralized Network & API Monitoring</p>
        </div>
    </header>

    <main class="container my-5">
        <div id="mainHub" class="row justify-content-center">
            <div class="col-md-6">
                <div class="card p-4 text-center">
                    <h3 class="mb-3 text-info"><i class="fas fa-users me-2"></i> Guest Access</h3>
                    <p class="text-muted small">Enter your designation to initiate a secure connection to the network grid.</p>

                    <div class="mb-3 text-start">
                        <label class="form-label text-pure-white">Name</label>
                        <input type="text" id="visitorName" class="form-control" placeholder="[e.g., NetRunner_01]">
                    </div>
                    <button onclick="navigateToHomePage()" class="btn btn-info w-100">Jump In!</button>
                </div>
            </div>
        </div>

        <div id="homePageContent" class="row justify-content-center d-none page-transition">
            <div class="col-md-10">
                <div class="card p-5">
                    <h2 class="text-info mb-4">Welcome to Marcell's Digital World</h2>
                    <p class="text-muted">Your presence has been registered. Explore the core systems below.</p>

                    <ul class="nav nav-tabs mb-4" id="myTab" role="tablist">
                        <li class="nav-item" role="presentation">
                            <button class="nav-link active" id="discussion-tab" data-bs-toggle="tab" data-bs-target="#discussion" type="button" role="tab" aria-controls="discussion" aria-selected="true">Discussion Log</button>
                        </li>
                        <li class="nav-item" role="presentation">
                            <button class="nav-link" id="gallery-tab" data-bs-toggle="tab" data-bs-target="#gallery" type="button" role="tab" aria-controls="gallery" aria-selected="false">Gallery</button>
                        </li>
                        <li class="nav-item" role="presentation">
                            <button class="nav-link" id="links-tab" data-bs-toggle="tab" data-bs-target="#links" type="button" role="tab" aria-controls="links" aria-bs-selected="false">Links</button>
                        </li>
                    </ul>
                    <div class="tab-content" id="myTabContent">
                        <div class="tab-pane fade show active" id="discussion" role="tabpanel" aria-labelledby="discussion-tab">
                            <div class="mb-3">
                                <label class="form-label text-pure-white visually-hidden">Search Discussion</label>
                                <input type="text" id="discussionSearchInput" class="form-control" placeholder="Search discussion log entries..." onkeyup="fetchDiscussionLogs()">
                            </div>
                            <div class="p-3 rounded my-4" style="background-color: #0d0d1a; border: 1px solid var(--border-color);">
                                <div id="discussionLogContent">
                                    <p class="mb-0 text-muted font-monospace">// Initializing discussion log interface...</p>
                                </div>
                            </div>
                            <button class="btn btn-outline-info btn-sm mt-2" onclick="fetchDiscussionLogs()"><i class="fas fa-sync-alt me-1"></i> Refresh Log</button>
                            <div class="mt-3">
                                <h6 class="text-pure-white mb-2">Leave a Comment:</h6>
                                <textarea id="commentInput" class="form-control mb-2" rows="2" placeholder="Share your thoughts..."></textarea>
                                <button class="btn btn-outline-info btn-sm" onclick="submitComment()">Submit Comment</button>
                                <div id="commentStatus" class="alert d-none mt-2"></div>
                            </div>
                        </div>
                        <div class="tab-pane fade" id="gallery" role="tabpanel" aria-labelledby="gallery-tab">
                            <div class="p-3 rounded my-4" style="background-color: #0d0d1a; border: 1px solid var(--border-color);">
                                <p class="mb-0 text-muted">[Image Placeholder: Cybernetic Landscapes]</p>
                                <p class="mb-0 text-muted">[Image Placeholder: Digital Architecture]</p>
                            </div>
                        </div>
                        <div class="tab-pane fade" id="links" role="tabpanel" aria-labelledby="links-tab">
                            <div class="p-3 rounded my-4" style="background-color: #0d0d1a; border: 1px solid var(--border-color);">
                                <div id="socialMediaLinksDisplay">
                                    <p class="mb-0 text-muted">// No links set yet. Check admin panel.</p>
                                </div>
                            </div>
                        </div>
                    </div>

                    <button onclick="returnToHub()" class="btn btn-outline-secondary btn-sm mt-4">Disconnect & Return</button>
                </div>
            </div>
        </div>

        <div id="adminDashboard" class="row d-none">
            <div class="col-12">
                <div class="card p-4">
                    <div class="d-flex justify-content-between align-items-center border-bottom pb-3 mb-3" style="border-color: var(--border-color) !important;">
                        <h3 class="text-warning mb-0"><i class="fas fa-server me-2"></i> Administrative Console // System Logs</h3>
                        <div>
                            <button onclick="refreshLogs()" class="btn btn-sm btn-outline-warning me-2">Refresh Log Stream</button>
                            <button onclick="logoutAdmin()" class="btn btn-sm btn-outline-danger">Terminate Session</button>
                        </div>
                    </div>

                    <ul class="nav nav-tabs mb-4" id="adminTabs" role="tablist">
                        <li class="nav-item" role="presentation">
                            <button class="nav-link active" id="admin-logs-tab" data-bs-toggle="tab" data-bs-target="#admin-logs" type="button" role="tab" aria-controls="admin-logs" aria-selected="true"><i class="fas fa-chart-line me-1"></i> Activity Logs</button>
                        </li>
                        <li class="nav-item" role="presentation">
                            <button class="nav-link" id="admin-gallery-tab" data-bs-toggle="tab" data-bs-target="#admin-gallery" type="button" role="tab" aria-controls="admin-gallery" aria-selected="false"><i class="fas fa-images me-1"></i> Gallery Management</button>
                        </li>
                        <li class="nav-item" role="presentation">
                            <button class="nav-link" id="admin-links-tab" data-bs-toggle="tab" data-bs-target="#admin-links" type="button" role="tab" aria-controls="admin-links" aria-selected="false"><i class="fas fa-link me-1"></i> Social Media Links</button>
                        </li>
                    </ul>

                    <div class="tab-content" id="adminTabContent">
                        <div class="tab-pane fade show active" id="admin-logs" role="tabpanel" aria-labelledby="admin-logs-tab">
                            <div class="mb-3">
                                <p class="text-pure-white mb-1">Total Unique Visitors: <span id="uniqueVisitorsCount" class="text-info">0</span></p>
                                <p class="text-pure-white">Total API Calls: <span id="totalApiCalls" class="text-info">0</span></p>
                            </div>
                            <pre id="logDisplay" class="p-3" style="max-height: 400px; overflow-y: auto;">AWAITING SYSTEM TELEMETRY RECORDS...</pre>
                        </div>
                        <div class="tab-pane fade" id="admin-gallery" role="tabpanel" aria-labelledby="admin-gallery-tab">
                            <h5 class="text-info mb-3">Upload New Gallery Item</h5>
                            <div class="input-group mb-3">
                                <input type="file" class="form-control" id="galleryImageUpload" accept="image/*">
                                <button class="btn btn-info" type="button" onclick="uploadGalleryImage()">Upload</button>
                            </div>
                            <div id="galleryUploadStatus" class="alert d-none"></div>

                            <h5 class="text-info mt-4 mb-3">Current Gallery Items</h5>
                            <div id="currentGalleryItems" class="row g-2">
                                <p class="text-muted">No images uploaded yet.</p>
                            </div>
                        </div>
                        <div class="tab-pane fade" id="admin-links" role="tabpanel" aria-labelledby="admin-links-tab">
                            <h5 class="text-info mb-3">Manage Social Media Links</h5>
                            <div class="mb-3">
                                <label class="form-label text-pure-white">GitHub URL</label>
                                <input type="url" class="form-control" id="githubLink" placeholder="https://github.com/yourprofile">
                            </div>
                            <div class="mb-3">
                                <label class="form-label text-pure-white">LinkedIn URL</label>
                                <input type="url" class="form-control" id="linkedinLink" placeholder="https://linkedin.com/in/yourprofile">
                            </div>
                            <div class="mb-3">
                                <label class="form-label text-pure-white">X (Twitter) URL</label>
                                <input type="url" class="form-control" id="twitterLink" placeholder="https://x.com/yourprofile">
                            </div>
                            <button class="btn btn-warning w-100" onclick="saveSocialMediaLinks()"><i class="fas fa-save me-1"></i> Save Links</button>
                            <div id="socialMediaStatus" class="alert d-none mt-3"></div>

                            <h5 class="text-info mt-4 mb-3">Current Social Media Links</h5>
                            <div id="currentSocialMediaLinks">
                                <p class="text-muted">No links set yet.</p>
                            </div>
                        </div>
                    </div>
                </div>
            </div>
        </div>
    </main>

    <div class="modal fade" id="settingsModal" tabindex="-1" aria-hidden="true">
        <div class="modal-dialog modal-dialog-centered">
            <div class="modal-content card p-2">
                <div class="modal-header border-0">
                    <h5 class="modal-title text-info"><i class="fas fa-cog me-2"></i> Interface Settings</h5>
                    <button type="button" class="btn-close" data-bs-dismiss="modal" aria-label="Close"></button>
                </div>
                <div class="modal-body">

                    <div class="mb-4 p-3 border" style="border-color: var(--theme-color) !important;">
                        <h6 class="text-pure-white mb-3">Theme Protocol</h6>
                        <div class="d-flex justify-content-between">
                            <div>
                                <input class="form-check-input" type="radio" name="themeColor" id="themeCyan" value="#00ffff" checked onchange="setTheme(this.value)">
                                <label class="form-check-label ms-2 text-pure-white" for="themeCyan">CYAN BLU</label>
                            </div>
                            <div>
                                <input class="form-check-input" type="radio" name="themeColor" id="themeMagenta" value="#ff00ff" onchange="setTheme(this.value)">
                                <label class="form-check-label ms-2 text-pure-white" for="themeMagenta">MAGENTA</label>
                            </div>
                            <div>
                                <input class="form-check-input" type="radio" name="themeColor" id="themeLime" value="#39ff14" onchange="setTheme(this.value)">
                                <label class="form-check-label ms-2 text-pure-white" for="themeLime">LIME GRN</label>
                            </div>
                        </div>
                    </div>

                    <div class="form-check form-switch mb-3 p-3 border" style="border-color: var(--theme-color) !important;">
                        <input class="form-check-input" type="checkbox" id="negativeViewToggle" onchange="toggleNegativeView()">
                        <label class="form-check-label ms-2 text-pure-white" for="negativeViewToggle">NEGATIVE VIEW (INVERT)</label>
                    </div>

                </div>
                <div class="modal-footer border-0">
                    <button type="button" class="btn btn-outline-secondary" data-bs-dismiss="modal">Close</button>
                </div>
            </div>
        </div>
    </div>

    <div class="modal fade" id="loginModal" tabindex="-1" aria-hidden="true">
        <div class="modal-dialog modal-dialog-centered">
            <div class="modal-content card p-2">
                <div class="modal-header border-0">
                    <h5 class="modal-title text-warning"><i class="fas fa-shield-alt me-2"></i> Security Protocol: ADMIN</h5>
                    <button type="button" class="btn-close" data-bs-dismiss="modal" aria-label="Close"></button>
                </div>
                <div class="modal-body">
                    <div class="mb-3">
                        <label class="form-label text-pure-white">Encryption Key</label>
                        <input type="password" id="adminPassword" class="form-control" placeholder="••••••••">
                    </div>
                    <div id="adminStatus" class="alert d-none mb-0"></div>
                </div>
                <div class="modal-footer border-0">
                    <button type="button" class="btn btn-outline-secondary" data-bs-dismiss="modal">Cancel</button>
                    <button type="button" onclick="checkAdminLogin()" class="btn btn-warning w-100">Authenticate</button>
                </div>
            </div>
        </div>
    </div>

    <footer class="container text-center text-muted small mt-5 pt-4 border-top">
        <p>&copy; 2077 Jacorian Marcell. All Rights Reserved. // Unauthorized access is a violation of NetWatch Protocols. Legal disclaimers apply.</p>
    </footer>

    <script src="https://cdn.jsdelivr.net/npm/bootstrap@5.3.0/dist/js/bootstrap.bundle.min.js"></script>

    <script>
        let currentVisitorName = 'Anonymous NetRunner'; // Global to store visitor's name

        function setTheme(color) {
            document.documentElement.style.setProperty('--theme-color', color);
            // Update glow color based on theme color
            let glowColor = color;
            if (color === '#00ffff') glowColor = 'rgba(0, 255, 255, 0.6)';
            else if (color === '#ff00ff') glowColor = 'rgba(255, 0, 255, 0.6)';
            else if (color === '#39ff14') glowColor = 'rgba(57, 255, 20, 0.6)';
            document.documentElement.style.setProperty('--glow-color', glowColor);
        }

        function toggleNegativeView() {
            document.body.classList.toggle('negative-mode');
        }

        function navigateToHomePage() {
            currentVisitorName = document.getElementById('visitorName').value || 'Anonymous NetRunner';
            document.getElementById('mainHub').classList.add('d-none');
            const homePage = document.getElementById('homePageContent');
            homePage.classList.remove('d-none');
            setTimeout(() => homePage.classList.add('active'), 10);
            google.colab.kernel.invokeFunction('notebook.log_visit', [currentVisitorName], {});
        }

        function returnToHub() {
            document.getElementById('visitorName').value = '';
            const homePage = document.getElementById('homePageContent');
            homePage.classList.remove('active');
            setTimeout(() => {
                homePage.classList.add('d-none');
                document.getElementById('mainHub').classList.remove('d-none');
            }, 500); // Match CSS transition duration
        }

        function checkAdminLogin() {
            const password = document.getElementById('adminPassword').value;
            google.colab.kernel.invokeFunction('notebook.admin_login', [password], {});
        }

        function showAdminDashboard(logsJson, uniqueVisitorsCount) {
            const modalEl = document.getElementById('loginModal');
            const modal = bootstrap.Modal.getInstance(modalEl);
            if(modal) modal.hide();

            document.getElementById('mainHub').classList.add('d-none');
            document.getElementById('homePageContent').classList.add('d-none'); // Hide homepage too if admin logs in from there
            document.getElementById('adminDashboard').classList.remove('d-none');

            updateAdminLogs(logsJson, uniqueVisitorsCount);
            fetchGalleryImages(); // Fetch gallery images when admin dashboard is shown
            fetchSocialMediaLinks(); // Fetch social media links when admin dashboard is shown
        }

        function updateAdminLogs(logsJson, uniqueVisitorsCount) {
            document.getElementById('logDisplay').innerText = logsJson;
            document.getElementById('uniqueVisitorsCount').innerText = uniqueVisitorsCount;
            document.getElementById('totalApiCalls').innerText = JSON.parse(logsJson).length; // Assuming logsJson is array of calls
        }

        function denyAdminAccess() {
            const errDiv = document.getElementById('adminStatus');
            errDiv.className = "alert alert-danger mt-2"; /* Simplified for cyberpunk alert */
            errDiv.innerText = "ACCESS DENIED: Unauthorized Encryption Key.";
            errDiv.classList.remove('d-none');
        }

        function logoutAdmin() {
            document.getElementById('adminPassword').value = '';
            document.getElementById('adminStatus').classList.add('d-none');
            document.getElementById('adminDashboard').classList.add('d-none');
            document.getElementById('mainHub').classList.remove('d-none');
        }

        function refreshLogs() {
            google.colab.kernel.invokeFunction('notebook.get_admin_logs', [], {});
        }

        function fetchDiscussionLogs() {
            const searchQuery = document.getElementById('discussionSearchInput').value;
            google.colab.kernel.invokeFunction('notebook.get_discussion_logs', [searchQuery], {})
                .then(function(result) {
                    const logContentDiv = document.getElementById('discussionLogContent');
                    logContentDiv.innerHTML = ''; // Clear previous logs
                    if (result && result.length > 0) {
                        result.forEach((log, index) => {
                            const logEntryDiv = document.createElement('div');
                            logEntryDiv.className = 'pb-3 mb-3 border-bottom'; // Separator for entries
                            logEntryDiv.innerHTML = `
                                <p class="mb-1 text-muted font-monospace d-flex justify-content-between align-items-center">
                                    // [${log.timestamp}] ${log.entry}
                                    <span class="d-flex align-items-center">
                                        <button class="btn btn-outline-info btn-sm me-2 py-0 px-1" onclick="addReaction(${index}, 'like')"><i class="fas fa-thumbs-up"></i> <span class="ms-1">${log.reactions.like || 0}</span></button>
                                        <button class="btn btn-outline-danger btn-sm me-2 py-0 px-1" onclick="addReaction(${index}, 'heart')"><i class="fas fa-heart"></i> <span class="ms-1">${log.reactions.heart || 0}</span></button>
                                    </span>
                                </p>
                                <div class="comments-section ms-3 ps-2 border-start" style="border-color: rgba(0, 255, 255, 0.3) !important;">
                                    <!-- Comments will be appended here -->
                                </div>
                            `;

                            const commentsSection = logEntryDiv.querySelector('.comments-section');
                            if (log.comments && log.comments.length > 0) {
                                log.comments.forEach(comment => {
                                    const commentP = document.createElement('p');
                                    commentP.className = 'mb-1 small text-secondary font-monospace';
                                    commentP.innerHTML = `// [${comment.timestamp}] &lt;${comment.author}&gt;: ${comment.text}`;
                                    commentsSection.appendChild(commentP);
                                });
                            } else {
                                const noCommentP = document.createElement('p');
                                noCommentP.className = 'mb-1 small text-secondary font-monospace';
                                noCommentP.innerText = '// No comments yet.';
                                commentsSection.appendChild(noCommentP);
                            }
                            logContentDiv.appendChild(logEntryDiv);
                        });
                    } else {
                        logContentDiv.innerHTML = '<p class="mb-0 text-muted font-monospace">// No discussion entries found matching your search.</p>';
                    }
                });
        }

        function addReaction(entryId, reactionType) {
            google.colab.kernel.invokeFunction('notebook.track_reaction', [entryId, reactionType], {})
                .then(() => fetchDiscussionLogs()); // Refresh logs to show updated counts
        }

        function submitComment() {
            const commentText = document.getElementById('commentInput').value;
            if (!commentText.trim()) {
                displayCommentStatus('Comment cannot be empty.', 'danger');
                return;
            }
            // For simplicity, associating comments with the latest discussion entry for now.
            // A more robust system would involve selecting which entry to comment on.
            google.colab.kernel.invokeFunction('notebook.add_discussion_comment', [currentVisitorName, commentText], {})
                .then(() => {
                    document.getElementById('commentInput').value = ''; // Clear input
                    displayCommentStatus('Comment submitted!', 'success');
                    fetchDiscussionLogs(); // Refresh logs to show new comment
                });
        }

        function displayCommentStatus(message, type) {
            const statusDiv = document.getElementById('commentStatus');
            statusDiv.className = `alert alert-${type} mt-2`;
            statusDiv.innerText = message;
            statusDiv.classList.remove('d-none');
            setTimeout(() => statusDiv.classList.add('d-none'), 3000); // Hide after 3 seconds
        }

        function uploadGalleryImage() {
            const fileInput = document.getElementById('galleryImageUpload');
            if (fileInput.files.length === 0) {
                displayGalleryUploadStatus('Please select a file to upload.', 'danger');
                return;
            }
            const file = fileInput.files[0];
            const reader = new FileReader();

            reader.onload = function(e) {
                const base64Data = e.target.result.split(',')[1]; // Get base64 string after comma
                google.colab.kernel.invokeFunction('notebook.upload_gallery_image', [file.name, base64Data], {});
            };
            reader.readAsDataURL(file);
        }

        function displayGalleryUploadStatus(message, type) {
            const statusDiv = document.getElementById('galleryUploadStatus');
            statusDiv.className = `alert alert-${type} mt-2`;
            statusDiv.innerText = message;
            statusDiv.classList.remove('d-none');
            setTimeout(() => statusDiv.classList.add('d-none'), 3000); // Hide after 3 seconds
        }

        function fetchGalleryImages() {
            google.colab.kernel.invokeFunction('notebook.get_gallery_images', [], {})
                .then(function(images) {
                    const galleryDiv = document.getElementById('currentGalleryItems');
                    galleryDiv.innerHTML = '';
                    if (images && images.length > 0) {
                        images.forEach(image => {
                            const colDiv = document.createElement('div');
                            colDiv.className = 'col-auto';
                            const img = document.createElement('img');
                            img.src = `data:image/jpeg;base64,${image.data}`; // Assuming JPEG and storing full base64
                            img.alt = image.filename;
                            img.className = 'gallery-image-preview';
                            colDiv.appendChild(img);
                            galleryDiv.appendChild(colDiv);
                        });
                    } else {
                        galleryDiv.innerHTML = '<p class="text-muted">No images uploaded yet.</p>';
                    }
                });
        }

        function saveSocialMediaLinks() {
            const github = document.getElementById('githubLink').value;
            const linkedin = document.getElementById('linkedinLink').value;
            const twitter = document.getElementById('twitterLink').value;
            const links = { github, linkedin, twitter };
            google.colab.kernel.invokeFunction('notebook.save_social_media_links', [links], {});
        }

        function displaySocialMediaStatus(message, type) {
            const statusDiv = document.getElementById('socialMediaStatus');
            statusDiv.className = `alert alert-${type} mt-3`;
            statusDiv.innerText = message;
            statusDiv.classList.remove('d-none');
            setTimeout(() => statusDiv.classList.add('d-none'), 3000); // Hide after 3 seconds
        }

        function fetchSocialMediaLinks() {
            google.colab.kernel.invokeFunction('notebook.get_social_media_links', [], {})
                .then(function(links) {
                    const linksDiv = document.getElementById('currentSocialMediaLinks');
                    linksDiv.innerHTML = '';
                    if (links) {
                        document.getElementById('githubLink').value = links.github || '';
                        document.getElementById('linkedinLink').value = links.linkedin || '';
                        document.getElementById('twitterLink').value = links.twitter || '';

                        if (links.github) linksDiv.innerHTML += `<p class="mb-1"><a href="${links.github}" target="_blank" class="text-info"><i class="fab fa-github me-1"></i> GitHub</a></p>`;
                        if (links.linkedin) linksDiv.innerHTML += `<p class="mb-1"><a href="${links.linkedin}" target="_blank" class="text-info"><i class="fab fa-linkedin me-1"></i> LinkedIn</a></p>`;
                        if (links.twitter) linksDiv.innerHTML += `<p class="mb-1"><a href="${links.twitter}" target="_blank" class="text-info"><i class="fab fa-twitter me-1"></i> X (Twitter)</a></p>`;

                        if (!links.github && !links.linkedin && !links.twitter) {
                            linksDiv.innerHTML = '<p class="text-muted">No links set yet.</p>';
                        }
                    }
                     // Update frontend display links based on fetched data
                    const homepageLinksDiv = document.getElementById('socialMediaLinksDisplay');
                    homepageLinksDiv.innerHTML = '';
                    if (links && (links.github || links.linkedin || links.twitter)) {
                        if (links.github) homepageLinksDiv.innerHTML += `<p class="mb-0"><a href="${links.github}" target="_blank" class="text-info">// GITHUB REPOSITORY [EXTERNAL LINK]</a></p>`;
                        if (links.linkedin) homepageLinksDiv.innerHTML += `<p class="mb-0"><a href="${links.linkedin}" target="_blank" class="text-info">// LINKEDIN PROFILE [EXTERNAL LINK]</a></p>`;
                        if (links.twitter) homepageLinksDiv.innerHTML += `<p class="mb-0"><a href="${links.twitter}" target="_blank" class="text-info">// X (TWITTER) PROFILE [EXTERNAL LINK]</a></p>`;
                    } else {
                        homepageLinksDiv.innerHTML = '<p class="mb-0 text-muted">// No links set yet. Check admin panel.</p>';
                    }
                });
        }


        // Add event listener for 'Enter' key on visitorName input
        document.addEventListener('DOMContentLoaded', (event) => {
            const visitorNameInput = document.getElementById('visitorName');
            if (visitorNameInput) {
                visitorNameInput.addEventListener('keypress', function (e) {
                    if (e.key === 'Enter') {
                        navigateToHomePage();
                    }
                });
            }
            // Automatically fetch logs when the homepage is activated and the discussion tab is shown
            const discussionTabButton = document.getElementById('discussion-tab');
            if (discussionTabButton) {
                discussionTabButton.addEventListener('shown.bs.tab', function (e) {
                    fetchDiscussionLogs();
                });
            }

            // Event listeners for admin tabs
            const adminLogsTab = document.getElementById('admin-logs-tab');
            if (adminLogsTab) {
                adminLogsTab.addEventListener('shown.bs.tab', function (e) {
                    refreshLogs(); // Refresh logs when logs tab is shown
                });
            }

            const adminGalleryTab = document.getElementById('admin-gallery-tab');
            if (adminGalleryTab) {
                adminGalleryTab.addEventListener('shown.bs.tab', function (e) {
                    fetchGalleryImages(); // Fetch images when gallery tab is shown
                });
            }

            const adminLinksTab = document.getElementById('admin-links-tab');
            if (adminLinksTab) {
                adminLinksTab.addEventListener('shown.bs.tab', function (e) {
                    fetchSocialMediaLinks(); // Fetch links when social media tab is shown
                });
            }

            // Global click listener for spark effect
            document.body.addEventListener('click', function(e) {
                createSpark(e.clientX, e.clientY);
            });
        });

        function createSpark(x, y) {
            const spark = document.createElement('div');
            spark.className = 'spark';
            spark.style.left = `${x}px`;
            spark.style.top = `${y}px`;
            document.body.appendChild(spark);

            spark.addEventListener('animationend', () => {
                spark.remove();
            });
        }

    </script>
</body>
</html>
"""

with open('index.html', 'w') as f:
    f.write(html_content)

# ==========================================
# 2. THE BACKEND TRACKER LOGIC (PYTHON)
# ==========================================
class APITracker:
    def __init__(self):
        self.calls = []
        self.discussion_entries = [
            {"timestamp": datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S"), "entry": "SYSTEM: Initiating data stream protocols...", "reactions": {"like": 0, "heart": 0}, "comments": []},
            {"timestamp": datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S"), "entry": "USER: 'Anonymous NetRunner' has connected.", "reactions": {"like": 0, "heart": 0}, "comments": []},
            {"timestamp": datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S"), "entry": "ALERT: Firewall integrity check complete. All systems nominal.", "reactions": {"like": 0, "heart": 0}, "comments": []}
        ]
        self.unique_visitors = set() # To track unique visitors for "reach"
        self.gallery_images = [] # Store filenames and base64 data
        self.social_media_links = {
            "github": "",
            "linkedin": "",
            "twitter": ""
        }

    def log_request(self, endpoint, method, payload=None):
        request_data = {
            "timestamp": datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S"),
            "type": "INBOUND_REQUEST",
            "endpoint": endpoint,
            "method": method,
            "payload": payload
        }
        self.calls.append(request_data)
        return request_data

    def log_response(self, request_data, status_code, response_body=None):
        response_data = {
            "timestamp": datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S"),
            "type": "OUTBOUND_RESPONSE",
            "request_id": request_data.get("timestamp"),
            "status_code": status_code,
            "response_body": response_body,
        }
        self.calls.append(response_data)

    def get_discussion_entries(self, search_query=None):
        # Simulate new real-time entries for demonstration
        if len(self.discussion_entries) < 6:
            self.discussion_entries.append({
                "timestamp": datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S"),
                "entry": f"SYSTEM: New data packet detected. ID: {len(self.discussion_entries) + 1}",
                "reactions": {"like": 0, "heart": 0},
                "comments": []
            })

        if search_query:
            search_query_lower = search_query.lower()
            return [entry for entry in self.discussion_entries if search_query_lower in entry['entry'].lower()]
        return self.discussion_entries

    def track_visitor(self, name):
        self.unique_visitors.add(name)

    def get_unique_visitors_count(self):
        return len(self.unique_visitors)

    def upload_gallery_image(self, filename, base64_data):
        image_info = {
            "id": len(self.gallery_images) + 1,
            "filename": filename,
            "data": base64_data, # Store full base64 data for display
            "timestamp": datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")
        }
        self.gallery_images.append(image_info)
        return True # Indicate success

    def get_gallery_images(self):
        return self.gallery_images

    def update_social_media_links(self, links):
        self.social_media_links.update(links)
        return True # Indicate success

    def get_social_media_links(self):
        return self.social_media_links

    def track_reaction(self, entry_id, reaction_type):
        # Ensure entry_id is within bounds and has reactions field
        if 0 <= entry_id < len(self.discussion_entries) and 'reactions' in self.discussion_entries[entry_id]:
            if reaction_type in self.discussion_entries[entry_id]['reactions']:
                self.discussion_entries[entry_id]['reactions'][reaction_type] += 1
                return True
        return False

    def add_comment(self, author, text):
        # For now, comments are added to the *latest* discussion entry.
        # A more complex system would allow commenting on specific entries.
        if self.discussion_entries:
            self.discussion_entries[-1]['comments'].append({
                "timestamp": datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S"),
                "author": author,
                "text": text
            })
            return True
        return False

# Boot track stream
tracker = APITracker()
ADMIN_PASSWORD = "Zillathepope"

def log_visit(name):
    # Logs the exact user login name directly inside the tracking records database array
    tracker.track_visitor(name)
    req_log = tracker.log_request("/hub/entry", "POST", payload={"tracked_username": name})
    tracker.log_response(req_log, 200, response_body={"status": "session_captured", "user": name})

def admin_login(password):
    if password == ADMIN_PASSWORD:
        output.eval_js(f"showAdminDashboard(JSON.stringify({tracker.calls}, null, 2), {tracker.get_unique_visitors_count()})")
    else:
        output.eval_js("denyAdminAccess()")

def get_admin_logs():
    output.eval_js(f"updateAdminLogs(JSON.stringify({tracker.calls}, null, 2), {tracker.get_unique_visitors_count()})")

def get_discussion_logs_callback(search_query=None):
    # Returning a deep copy to prevent direct modification from JS and ensure consistent serialization
    return [entry.copy() for entry in tracker.get_discussion_entries(search_query)]

def upload_gallery_image_backend(filename, base64_data):
    success = tracker.upload_gallery_image(filename, base64_data)
    if success:
        output.eval_js("displayGalleryUploadStatus('Image uploaded successfully!', 'success'); fetchGalleryImages();")
    else:
        output.eval_js("displayGalleryUploadStatus('Image upload failed!', 'danger');")

def get_gallery_images_backend():
    images = tracker.get_gallery_images()
    output.eval_js(f"displayGalleryItems(JSON.stringify({images}))") # Keeping displayGalleryItems for consistency

def save_social_media_links_backend(links):
    success = tracker.update_social_media_links(links)
    if success:
        output.eval_js("displaySocialMediaStatus('Social media links saved!', 'success'); fetchSocialMediaLinks();")
    else:
        output.eval_js("displaySocialMediaStatus('Failed to save social media links!', 'danger');")

def get_social_media_links_backend():
    links = tracker.get_social_media_links()
    output.eval_js(f"displaySocialMediaLinks(JSON.stringify({links}))") # Keeping displaySocialMediaLinks for consistency

def track_reaction_backend(entry_id, reaction_type):
    success = tracker.track_reaction(entry_id, reaction_type)
    # No specific frontend message needed, as fetchDiscussionLogs will refresh counts

def add_discussion_comment_backend(author, text):
    success = tracker.add_comment(author, text)
    # Frontend will refresh logs to show the new comment.

# Connect Frontend to Backend bridges
output.register_callback('notebook.log_visit', log_visit)
output.register_callback('notebook.admin_login', admin_login)
output.register_callback('notebook.get_admin_logs', get_admin_logs)
output.register_callback('notebook.get_discussion_logs', get_discussion_logs_callback)
output.register_callback('notebook.upload_gallery_image', upload_gallery_image_backend)
output.register_callback('notebook.get_gallery_images', get_gallery_images_backend)
output.register_callback('notebook.save_social_media_links', save_social_media_links_backend)
output.register_callback('notebook.get_social_media_links', get_social_media_links_backend)
output.register_callback('notebook.track_reaction', track_reaction_backend)
output.register_callback('notebook.add_discussion_comment', add_discussion_comment_backend)

# ==========================================
# 3. LAUNCH INTERFACE
# ==========================================
print("All adjustments successfully executed. View layout rendered directly below:")
IPython.display.HTML(filename='index.html")
```
