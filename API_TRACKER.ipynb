```python
import datetime

class APITracker:
    def __init__(self):
        self.calls = []

    def log_request(self, endpoint, method, payload=None, headers=None):
        request_data = {
            "timestamp": datetime.datetime.now().isoformat(),
            "type": "request",
            "endpoint": endpoint,
            "method": method,
            "payload": payload,
            "headers": headers
        }
        self.calls.append(request_data)
        print(f"Logged request to {endpoint} ({method})")
        return request_data

    def log_response(self, request_data, status_code, response_body=None, response_headers=None):
        response_data = {
            "timestamp": datetime.datetime.now().isoformat(),
            "type": "response",
            "request_id": request_data.get("timestamp"),
            "status_code": status_code,
            "response_body": response_body,
            "response_headers": response_headers
        }
        self.calls.append(response_data)
        print(f"Logged response for {request_data.get('endpoint')} with status {status_code}")

    def get_all_calls(self):
        return self.calls

    def clear_logs(self):
        self.calls = []
        print("API tracker logs cleared.")

print("APITracker class defined successfully")
```

```python
import datetime
from google.colab import output
import IPython

class APITracker:
    def __init__(self):
        self.calls = []

    def log_request(self, endpoint, method, payload=None):
        request_data = {
            "timestamp": datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S"),
            "type": "INBOUND_REQUEST",
            "endpoint": endpoint,
            "method": method,
            "payload": payload
        }
        self.calls.append(request_data)
        return request_data

    def log_response(self, request_data, status_code, response_body=None):
        response_data = {
            "timestamp": datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S"),
            "type": "OUTBOUND_RESPONSE",
            "request_id": request_data.get("timestamp"),
            "status_code": status_code,
            "response_body": response_body,
        }
        self.calls.append(response_data)

tracker = APITracker()
ADMIN_PASSWORD = "Zillathepope"

def log_visit(name):
    req_log = tracker.log_request("/portfolio/visit", "POST", payload={"node_id": name})
    tracker.log_response(req_log, 200, response_body={"status": "online", "welcome": f"Hello {name}"})

def admin_login(password):
    if password == ADMIN_PASSWORD:
        output.eval_js(f"showAdminDashboard({tracker.calls})")
    else:
        output.eval_js("denyAdminAccess()")

def get_admin_logs():
    output.eval_js(f"document.getElementById('logDisplay').innerText = JSON.stringify({tracker.calls}, null, 2)")

output.register_callback('notebook.log_visit', log_visit)
output.register_callback('notebook.admin_login', admin_login)
output.register_callback('notebook.get_admin_logs', get_admin_logs)

print("Backend system initialized and ready!")
```
